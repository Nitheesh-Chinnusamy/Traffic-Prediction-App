```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Dataset.csv")
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 
```

```python
dataset['Date']= le.fit_transform(dataset['Date'])
```

```python
dataset.tail(10)
```

```python
dataset.info()
```

```python
X = dataset[["CodedDay", "Zone", "Temperature"]].values
y = dataset.iloc[:, 6:7].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 300, random_state = 0)
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
if(y_pred.all()<2.5):
    y_pred=np.round(y_pred-0.5)
else:
    y_pred=np.round(y_pred+0.5)
```

```python
df1=(y_pred-y_test)/y_test
df1=round(df1.mean()*100,2)
print("Error = ",df1,"%") 
a=100-df1
print("Accuracy= ",a,"%")
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
if(y_pred.all()<2.5):
    y_pred=np.round(y_pred-0.5)
    
else:
    y_pred=np.round(y_pred+0.5)

y_pred
```

```python
df1=(y_pred-y_test)/y_test
df1=round(df1.mean()*100,2)
print("Error = ",df1,"%") 
```

```python
a=100-df1
print("Accuracy= ",a,"%")
```

```python
print("Error = ",df1,"%")
print("Accuracy= ",a,"%")
# Save the final trained SVR model to a file
import joblib
joblib.dump(regressor, "traffic_model.pkl")
print("✅ Model saved as traffic_model.pkl")
```
